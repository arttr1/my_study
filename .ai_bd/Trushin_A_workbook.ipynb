# задача 4.3.1

функция random использует равномерное распределение, а значит медиана и среднее значение будут стремиться к 0.5 при данном промежутке

```python
from random import randint
from matplotlib import pyplot as plt
import numpy as np

arr = [randint(1, 999)/1000 for x in range(500)]

mean = np.mean(arr)
median = np.median(arr)

print(mean)
print(median)

plt.scatter(range(len(arr)), arr)
```

# задание 4.3.5


```python
import numpy as np
def calculator(expression):
    expression = expression.replace(' ', '')

    if expression.startswith("sin("):
        st_in = 4
        end_in = expression.find(")")
        args = expression[st_in:end_in]
        x, y = map(float, args.split("+"))
        return np.sin(np.deg2rad(x + y))
    if expression.startswith("cos("):
        st_in = 4
        end_in = expression.find(")") 
        args = expression[st_in:end_in]
        x, y = map(float, args.split("+"))
        return np.cos(np.deg2rad(x + y))

    elif '**' in expression:
        x, y = expression.split("**")
        if x == 'e':
            return np.pow(np.e, float(y))
        else:
            return np.pow(float(x), float(y))
    elif '+' in expression:
        x, y = map(float, expression.split('+'))
        return x + y
    elif '-' in expression:
        x, y = map(float, expression.split('-'))
        return x - y
    elif '*' in expression:
        x, y = map(float, expression.split('*'))
        return x * y
    elif '/' in expression:
        x, y = map(float, expression.split('/'))
        return x / y
    
expression = input("введите выражение: ")
calculator(expression)
```
